`[0.2]` Create a new column `whoWon` which contains `"red"` or `"blue"` based on the `blueWins` column.

````python
# Your code here
def whoWon(lol_data):
    if lol_data.blueWins == 0:
        return "red"
    else:
        return "blue"

lol_data['whoWon'] = lol_data.apply(whoWon, axis='columns')
lol_data
````

````python
# Your code here
def firstKill(lol_data):
    if lol_data.blueFirstBlood == 0:
        return "red"
    else:
        return "blue"

lol_data['firstKill'] = lol_data.apply(firstKill, axis='columns')
lol_data
````
